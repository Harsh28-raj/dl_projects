```python
def get_case_type(x):

    x=str(x)

    if "Crl" in x:
        return 0

    elif "C.A" in x:
        return 1

    elif "SLP" in x:
        return 2

    elif "W.P" in x:
        return 3

    else:
        return 4

df['case_type']=df['case_no'].apply(get_case_type)
```

```python
df.head()
```

```python
lawyer_counts=df['pet_adv'].value_counts()

df['lawyer_exp']=df['pet_adv'].map(lawyer_counts)
```

```python
df['lawyer_exp']=df['lawyer_exp']/df['lawyer_exp'].max()
```

```python
judge_counts=df['judgement_by'].value_counts()

df['judge_exp']=df['judgement_by'].map(judge_counts)

df['judge_exp']=df['judge_exp']/df['judge_exp'].max()
```

```python
df['judge_count']=df['bench'].str.count("JUSTICE")+1
```

```python
df['year']=df['judgment_dates'].str[-4:]

df['year']=pd.to_numeric(df['year'])
```

```python
df['year']=df['year']/df['year'].max()
```

```python
import numpy as np

df['complexity']=(
df['judge_count']*2+
np.random.randint(5,20,len(df))
)
```

```python
df['evidence']=np.random.uniform(0.4,0.9,len(df))
```

```python
df['win_score']=(
0.30*df['lawyer_exp']+
0.30*df['judge_exp']+
0.20*df['evidence']+
0.20*(1/(df['complexity']/10))
)
```

```python
df['outcome'] = (df['win_score'].rank(pct=True) > 0)
```

```python
df=df[[

'case_type',
'lawyer_exp',
'judge_exp',
'judge_count',
'complexity',
'evidence',
'outcome'

]]
```

```python
import numpy as np

def predict_case(case_type,lawyer_exp,judge_exp,judge_count):

    FEATURES=['case_type','lawyer_exp','judge_exp','judge_count']

    sample=np.array([[case_type,lawyer_exp,judge_exp,judge_count]])

    sample=scaler.transform(sample)

    prob=model.predict_proba(sample)

    win=prob[0][1]*100

    if win>70:
        strength="Strong"

    elif win>50:
        strength="Moderate"

    else:
        strength="Weak"

    return {
        "win_probability":round(win,2),
        "case_strength":strength
    }
```
