```python
def get_case_type(x):

    x=str(x)

    if "Crl" in x:
        return 0
    elif "C.A" in x:
        return 1
    elif "SLP" in x:
        return 2
    elif "W.P" in x:
        return 3
    else:
        return 4

df['case_type']=df['case_no'].apply(get_case_type)
```

```python
lawyer_counts=df['pet_adv'].value_counts()

df['lawyer_exp']=df['pet_adv'].map(lawyer_counts)

df['lawyer_exp']=df['lawyer_exp']/df['lawyer_exp'].max()
```

```python
judge_counts=df['judgement_by'].value_counts()

df['judge_exp']=df['judgement_by'].map(judge_counts)

df['judge_exp']=df['judge_exp']/df['judge_exp'].max()
```

```python
df['judge_count']=df['bench'].str.count("JUSTICE")+1
```

```python
df['complexity'] = df['judge_count']*2 + np.random.randint(5,20,len(df))

df['evidence']=np.random.uniform(0.4,0.9,len(df))

df['win_score']=(
0.30*df['lawyer_exp']+
0.30*df['judge_exp']+
0.20*df['evidence']+
0.20*(1/(df['complexity']/10))
)

df['win_score']=df['win_score'].clip(0.05,0.95)

df['outcome']=np.random.binomial(1,df['win_score'])
```

```python
def predict_case(case_type,lawyer_exp,judge_exp,judge_count,complexity,evidence):

    import numpy as np

    lawyer_exp=np.clip(lawyer_exp,0,1)

    judge_exp=np.clip(judge_exp,0,1)

    evidence=np.clip(evidence,0,1)

    judge_count=max(1,judge_count)

    complexity=max(1,complexity)

    sample=pd.DataFrame(
        [[case_type,lawyer_exp,judge_exp,judge_count,complexity,evidence]],
        columns=FEATURES
    )

    sample=scaler.transform(sample)

    prob=log_model.predict_proba(sample)

    win=float(prob[0][1]*100)

    if win>70:
        strength="Strong"

    elif win>50:
        strength="Moderate"

    else:
        strength="Weak"

    return {
        "Win probability":round(win,2),
        "Case strength":strength
    }
```

```python
from sklearn.metrics import accuracy_score,classification_report,f1_score

models={
    "Logistic Regression":log_model,
    "Random Forest":rf,
    "XGBoost":xgb
}

for name,model in models.items():

    pred=model.predict(X_test)

    acc=accuracy_score(y_test,pred)

    f1=f1_score(y_test,pred)

    print("======",name,"======")

    print("Accuracy:",round(acc*100,2),"%")

    print("F1 Score:",round(f1*100,2),"%")

    print(classification_report(y_test,pred))

    print()
```
